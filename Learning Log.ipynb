## Data Analysis:  
Data analysis is the collection, transformation, and organization of data in order to draw conclusions, make predictions, and drive informed decision-making. 

Data analysis can be comprised of 6 steps:
1. Ask - Ask the people who are asking you for preparing the data
2. Prepare - Collect data from the people on whom the result needs to be applied 
3. Process - Clean and make relevant changes to the collected data
4. Analyze - analyze data to make predictions
5. Share - Documenting everything that I have done so far and share the same with the leadership of the company
6. Act - Taking action as per the results from the data that was analyzed

## Decoding Job Description

Data analytics consultant  analyzes the systems and models for using data

Data specialist  organizes or converts data for use in databases or software systems

Operations analyst  analyzes data to assess the performance of business operations and workflows

# Problem-solving and effective questioning
## The Six Data analysis phases

Step 1. Ask

Points to consider during this phase:
a. Define the problem we are trying to resolve.
b. Understanding the expectations of the stakeholders.
c. Keeping an open communication line with the stakeholders.

Step 2. Prepare

a. What metrics do I need to measure.
b. What data to collect.
c. Finding data in the database.
d. Protecting the data.

Step 3. Process

a. Cleaning the data.
b. Looking for inacuracies.
c. Removing spaces using SQL queries and excel .
d. Removing repeated entries.

Step 4. Analyze

a. Performing computations.
b. Making changes into the data.
c. Combining data from different sources.

Step 5. Share 

Sharing the outcomes in charts or tables with the stakeholders.

Step 6. Act

Applying the findings of your research to get get better results from the process.

### Ensuring Data Integrity

Data Bias - data being manupulated due to incorrect data collection



1. Sampling Bias is a bias when the sample collected is not collected evenly from entire population.
2. Observation bias (research/experiment) is abias which arise when data is collected incorrectly during observation.
3. Confirmation bias The tendency to search for or interpret information in a way that validates pre-existing beliefs.
4. Interpretation bias is a bias which arise when people interpret the data as per old experience.
