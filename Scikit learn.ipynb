## 0. end to end scikit learn workflow

```python
1.
#getting the data ready

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h_d=pd.read_csv("heart-disease.csv")
h_d
```

```python
2.
#defining the features column and the target column

#features table

x=h_d.drop("target",axis=1)
x

# target column

y=h_d["target"]
y
```

```python
#this is a classification problem as we need to classify if someone has heart disease or not
# hence, we need to choose a model as per the usage
# choosing right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifier is a classification ML model. Its capable of learning patterns in data 
# and then classifying a sample aka row is one thing or other

clf=RandomForestClassifier() # instantiation

clf.get_params()
```

```python
3. 
# fitting the model to training data
from sklearn.model_selection import train_test_split # to divide the exisiting data into training and testing data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

```python
# now fitting the training data into the ml model
clf.fit(x_train,y_train)
```

```python
# making a prediction
y_label=np.array([35,1,2,140,200,1,1,155,1,1.3,2,1])
y_label.reshape(-1,1)
```

```python
y_predict=clf.predict(x_test)
y_predict
```

```python
y_test
```

```python
# 4. Evaluating the model on the training data and test data
clf.score(x_train,y_train)
```

```python
# The sklearn. metrics module implements several loss, score, and utility functions to "measure classification performance" .

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test,y_predict))
```

```python
confusion_matrix(y_test,y_predict)
```
